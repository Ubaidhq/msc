```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
np.random.seed(16002703)
```

# Time Series Generation

```python
T = 2000
y_t = np.zeros(2000)
y_t[0] = 100 
y_t[1] = 100
e_t = np.random.standard_normal(size = T)
```

```python

phi = 0.6
d = 0.025
theta = -0.4 
for i in range(2,2000):
    dy_ = y_t[i-1] - y_t[i-2]
    dy = 0.6*(dy_ - d) + e_t[i] + theta*e_t[i-1]
    y_t[i] = y_t[i-1] + dy
```

```python
plt.plot(y_t)
```

```python
X_train, X_test = y_t[:1400], y_t[len(y_t) - 600:]
```

# Trading Strategies

I am going to look at the following strategies: buy and hold, trend-following and mean-reversion. 

```python
cash = np.zeros(np.shape(y_t))
cash[0] = 10000
```

### Buy and hold

I will be using this as a benchmark against our 3 trading strategies.

```python
TV_bh = cash[0]*X_train/X_train[0]
```

## Strategy 1: Mean-reversion 

Implementing a mean-reversion strategy similar to the one shown in lectures.  We will use a time window of 2 days.  

```python
#TRADING: MEAN REVERSION 
t = np.linspace(0, len(X_train), len(X_train))
time_window = 2
cumsum = [0]
sigPriceBuy = []
sigPriceSell = []

ma = np.zeros(np.shape(X_train))

w = np.zeros(np.shape(X_train))
cash = np.zeros(np.shape(X_train))

cash[0] = 10000

for i, x in enumerate(X_train[:-1], 0):
    cumsum.append(cumsum[i] + x)
    ma[i] = x
    if i>=time_window:
        moving_ave = (cumsum[i] - cumsum[i-time_window])/(time_window)
        ma[i] = moving_ave
    
    if ma[i] == x:
        w[i+1] = w[i]
        cash[i+1] = cash[i]
        sigPriceBuy.append(np.nan)
        sigPriceSell.append(np.nan)
        
    if ma[i] > x: 
        w[i+1] = cash[i]/x  + w[i]
        cash[i+1] = 0
        sigPriceBuy.append(x)
        sigPriceSell.append(np.nan)
        
    if ma[i] < x:
        cash[i+1] = w[i]*x + cash[i]
        w[i+1] = 0
        sigPriceBuy.append(np.nan)
        sigPriceSell.append(x)

ma[i+1] = X_train[len(X_train)-1]

mr_strategy = [a*b for a,b in zip(w,X_train)]+ cash


plt.plot(TV_bh, label = "Buy and hold")
plt.plot(t, mr_strategy, label="Mean reversion, w = 2 days")
plt.legend(loc="lower left")
plt.show()
```

```python
plt.figure(figsize=(12.6,4.6))
plt.plot(t, X_train, label='Price', alpha=0.5)
plt.scatter(t, sigPriceBuy, label='Buy', marker = '^', color='g')
plt.plot(t, sigPriceSell, label='Sell', marker='v', color='r')
```

```python
#TRADING: TREND FOLLOWING 
    
time_window = 2
cumsum = [0]

ma = np.zeros(np.shape(X_train))

w = np.zeros(np.shape(X_train))
cash = np.zeros(np.shape(X_train))

cash[0] = 10000
                
for i, x in enumerate(X_train[:-1], 0):
    cumsum.append(cumsum[i] + x)
    ma[i] = x
    if i>=time_window:
        moving_ave = (cumsum[i] - cumsum[i-time_window])/(time_window)
        ma[i] = moving_ave
    
    if ma[i] == x:
        w[i+1] = w[i]
        cash[i+1] = cash[i]
    
    if ma[i] < x: 
        w[i+1] = cash[i]/x  + w[i]
        cash[i+1] = 0
        
    if ma[i] > x:
        cash[i+1] = w[i]*x + cash[i]
        w[i+1] = 0

ma[i+1] = X_train[len(X_train)-1]
        
tf_strategy = [a*b for a,b in zip(w,X_train)]+ cash
```

```python
plt.plot(TV_bh, label = "Buy and hold")
plt.plot(t, mr_strategy, label="Mean reversion, w = 2 days")
plt.plot(t, tf_strategy, label="Trend Following, w = 2 days")
plt.legend(loc="upper left")
plt.show()
```
